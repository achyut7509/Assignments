## QUESTION - 20 (Data _set: Cars.csv)

```python
cars_data = pd.read_csv('Cars.csv',sep=',')
cars_data
```

```python
cars_data.shape
```

```python
cars_data.info()
```

```python
(cars_data['MPG']>38).sum()
```

```python
probability_greater_than_38 = ((cars_data['MPG']>38).sum()/cars_data['MPG'].sum())
probability_greater_than_38
```

```python
(cars_data['MPG']<40).sum()
```

```python
probability_less_than_40 = ((cars_data['MPG']<40).sum()/cars_data['MPG'].sum())
probability_less_than_40
```

```python
probability_between_range_20_to_50=((cars_data.MPG > 20) & (cars_data.MPG <50)).sum()
probability_between_range_20_to_50
```

```python
probability_between_range_20_to_50/(cars_data['MPG'].sum())
```
